# Download Data

## Download Data

```python
DOWNLOAD_ROOT = 'https://raw.githubusercontent.com/ageron/handson-ml2/master/'
HOUSING_PATH = os.path.join('datasets', 'housing')
HOUSING_URL = DOWNLOAD_ROOT + 'datasets/housing/housing.tgz'
```

```python
def fetch_housing_data(housing_url=HOUSING_URL, housing_path=HOUSING_PATH):
    os.makedirs(housing_path, exist_ok=True)
    tgz_path = os.path.join(housing_path, 'housing.tgz')
    urllib.request.urlretrieve(housing_url, tgz_path)
    housing_tgz = tarfile.open(tgz_path)
    housing_tgz.extractall(path=housing_path)
    housing_tgz.close()
```

```python
fetch_housing_data()
```

## Load Data

```python
def load_housing_data(housing_path=HOUSING_PATH):
    csv_path = os.path.join(housing_path, 'housing.csv')
    return pd.read_csv(csv_path)
```

## Stratified Train test split

If the sample is not large enough, there may be sampling bias.

For example, if a population is 53.1% female and 46.9% male, if a sample of 1000 people were to be taken , it should include 531 females and 469 males. 

This is called stratified sampling

```python
housing['income_cat'] = pd.cut(housing['median_income'],
                              bins=[0., 1.5, 3.0, 4.5, 6., np.inf],
                              labels=[1, 2, 3, 4, 5])
```

```python
housing['income_cat'].hist();
```

```python
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)

for train_index, test_index in split.split(housing, housing['income_cat']):
    strat_train_set = housing.loc[train_index]
    strat_test_set = housing.loc[test_index]
```

```python
strat_test_set['income_cat'].value_counts() / len(strat_test_set)
```

## Impute missing total_bedrooms

### Estimators

An object that can estimate some parameters is an estimator (imputer is an estimator).

The estimation is performed by the 'fit()' method.

Any other parameter needed to guide the estimation process is a hyperparameter (the imputer's strategy eg. Median)

### Transformers

Some estimators can also transform a dataset. 

The transformation is performed by the 'transform()' method.

Transformers generally rely on the learnt parameters (like the imputer).

All transformers also have a convenience method, 'fit_transform()' that may be optimize and run faster than calling fit() and transform() separately.

### Predictors

Some estimators, given a dataset, can make predictions.

LinearRegression model is a predictor. 

They have the method 'predict()' which takes a dataset of new instances and returns a dataset or predictions.

They have a 'score()' method that measure the quality of the predictions.

```python
imputer = SimpleImputer(strategy='median')
```

#### Limitations to Ordinal Encoding

As the name suggests, encoding lists categories in an order, implying that there may be a connection between closer numbers, rather than numbers further away.

This is useful if the categories are a rating (eg. Excellent, Very Good, Ok, etc)

However, it is not so useful here where categories 0 and 4 are more similar than categories 0 and 1.

This is where One Hot Encoding comes in useful.

### One Hot Encoding

This creates one binary attribute per category.

#### Sparse Matrix

### Custom Transformers

For custom cleanups, or combining specific attributes, you need to write your own custom transformer.

This is done by creating a class and implementing three methods (fit(), transform() and fit_transform()).

```python
# These are the column indices for each attribute
rooms_ix, bedrooms_ix, population_ix, households_ix = 3, 4, 5, 6

class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def __init__(self, add_bedrooms_per_room=True):
        self.add_bedrooms_per_room = add_bedrooms_per_room
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        rooms_per_household = X[:, rooms_ix] / X[:, households_ix]
        population_per_household = X[:, population_ix] / X[:, households_ix]
        if self.add_bedrooms_per_room:
            bedrooms_per_room = X[:, bedrooms_ix] / X[:, rooms_ix]
            # np_c concatenates arrays along the second axis (usually the horizontal)
            return np.c_[X, rooms_per_household, population_per_household, bedrooms_per_room]
        else:
            return np.c_[X, rooms_per_household, population_per_household]
        
attr_adder = CombinedAttributesAdder(add_bedrooms_per_room=False)

# The .values here converts the DF to a NumPy array before passing it into the transformer method.
housing_extra_attribs = attr_adder.transform(housing.values)
        

        
            
```

## Feature Scaling

Machine Learning algorithms don't perform well when in the input numerical attributes have very different scales.

Scaling the target values is generally not required.

It is important that the scalers are only fit to the training data only, not the test data. Then they can be used to transofmr the training set and the test set, as well as any new data.

There are two common ways to get all attributes to have the same scale:

### Min-Max scaling (or Normalisation)

Values are shifted and rescaled so that they all lie between 0 and 1. (SciKit-Learn has MinMaxScaler with a feature range parameter to change the range)

### Standardization

Calculated differently, the scaled features will have a mean of 0.

It does not bound values to a specific range, which can be a problem for some algorithms (Neural Networks usually expect an input value of between 0 and 1).

The advantage is that Standardization is much less affected by outliers.

SciKit-Learn has StandardScaler for this.

## Transformation Pipelines

SciKit-Learn provides the 'Pipeline' class to help carry out sequences of transformations.

The Pipeline constructor takes a list of name/estimator pairs defining a sequence of steps.

All but the last estimator must be transformers (they must have a fit_transform() method)

```python
num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('attribs_adder', CombinedAttributesAdder()),
    ('std_scaler', StandardScaler()),
])
```

```python
housing_num_tr = num_pipeline.fit_transform(housing_num)
```

### Column Transformers

So far, the categorical columns and the numerical columns have been handled separately. 

A ColumnTransformer is a single transformer that is able to handle all columns, applying appropriate transformations to each column.

```python
num_attribs = list(housing_num)
cat_attribs = ['ocean_proximity']

full_pipeline = ColumnTransformer([
    ('num', num_pipeline, num_attribs),
    ('cat', OneHotEncoder(), cat_attribs),
])

full_pipeline
```

```python
housing_prepared = full_pipeline.fit_transform(housing)
```

# Select and Train a Model

```python
lin_reg = LinearRegression()
lin_reg.fit(housing_prepared, housing_labels)
```
